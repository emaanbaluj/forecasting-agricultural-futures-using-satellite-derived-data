```python
import os, ee, geemap
from mosaic_processing import MosaicProcessing, export_to_drive, export_all_weeks

os.environ["EE_PROJECT_ID"] = "comp3200-476211"
ee.Authenticate()
ee.Initialize(project=os.getenv('EE_PROJECT_ID'))
print("✅ Earth Engine initialised with hidden project ID!")
```

```python
aoi = ee.Geometry.Rectangle([-104.0, 36.0, -80.0, 49.0])

mp = MosaicProcessing(
    aoi=aoi,
    start_date="2021-04-01",
    end_date="2021-10-01",
    cloud_prob_max=40,
    bands=["B2","B3","B4","B8","B11","B12","SCL","MSK_CLDPRB"]
)

rgb_weeks = mp.process_rgb(reducer="median")
ndvi_weeks = mp.process_ndvi(reducer="median")

print("Weekly RGB composites:", rgb_weeks.size().getInfo())
print("Weekly NDVI composites:", ndvi_weeks.size().getInfo())
```

```python
m = geemap.Map(center=[41.5, -93.5], zoom=5)

first_rgb = ee.Image(rgb_weeks.first())
m.addLayer(first_rgb, {"min": 0, "max": 3000, "bands": ["B4_median","B3_median","B2_median"]},
           "First-week RGB (median)")

first_ndvi = ee.Image(ndvi_weeks.first())
m.addLayer(first_ndvi.select("NDVI_median"),
           {"min": 0, "max": 1, "palette": ["white","green"]},
           "First-week NDVI (median)")

m
```
